### Bernstein Vazirani Algorithm Example

```python
from qiskit import *
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator

import matplotlib.pyplot as plt
```

```python
# Ask user to input a number and simulate it is unknown

# Input number
number_to_guess = int(input("Enter an integer to guess: "))

ntgb = bin(number_to_guess)[2:]     # It's a string again, delete the "0b" in front
ntgb = ntgb[::-1]                   # Reverse the binary number to match bit positions

# Length of input
bin_length = len(ntgb)

# Convert to list to integers and prepare number of qubits
number_to_guess_binary = []
qubit_number = []
for ix, i in enumerate(ntgb):
    number_to_guess_binary.append(int(i))
    qubit_number.append(int(ix))
```

```python
# Quantum Circuit

circuit = QuantumCircuit(bin_length+1, bin_length)

circuit.h(qubit_number)
circuit.x(bin_length)
circuit.h(bin_length)

for ix, i in enumerate(number_to_guess_binary):
    if i and 1:
        circuit.cx(ix, bin_length)

circuit.h(qubit_number)

circuit.measure(qubit_number, qubit_number)

circuit.draw(output='mpl')
```

```python
# Simulation, guess the number in only one try

simulator = AerSimulator(shots=1)

# Result and Plot
result = simulator.run(circuit).result()
counts = result.get_counts(circuit)

#print(counts.int_outcomes().keys())
# CHECK how to read values from counts.int_outcomes().keys(), data type = dict_keys([])
for i in counts.int_outcomes().keys():
    print (f"The secret number is: {i}")

```
